## 각 지번주소의 동별 cctv 개수를 Bar로 시각화하기

Figure_2.png

문자열 분리해주는 split 메서드

```python
exam = '마포동 100-45'
splited = exam.split()
print(splited)
print(splited[0])
print(splited[1])
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
read_lines = []
with open('마포구_cctv.csv') as f:
    reader = csv.reader(f)
    for line in reader:
        read_lines.append(line)

addresses = []
dongs = []
dongs_count = []
for line in read_lines[1:]: # 첫번째 헤더정보를 제외한 두번째줄부터 읽기
    if line[2]:
        address = line[2].split()[0] # 주소를 공백을 기준으로 나누고 앞쪽 정보만 담기
        addresses.append(address)

for address in addresses:
    if address not in dongs:
        count = addresses.count(address)  # 전체자료에서 해당 동의 개수 세기
        dongs.append(address)
        dongs_count.append(count)
```

데이터 이상치 처리
- 공백으로 나누면 이상치 발생 (사람의 실수로 인해 공백없이 주소를 넣거나 등등)

의도한 것과 다른 원본데이터에 의한 이상치를 어떻게 처리할 것인지 규칙 필요  
정확한 값으로 수정하거나  
이상치를 제거해 버리거나  
기타 등등 방법 사용

아래코드는 잘못된 자료는 제거하는 방법을 선택

```python
addresses = []
dongs = []
dongs_count = []
for line in read_lines[1:]:
    if line[2]:
        address = line[2].split()[0]
        addresses.append(address)

for address in addresses:
    if address not in dongs:
        count = addresses.count(address)
        if count > 10:  # 개수가 10개 미만인 이상치는 추가하지않고 제거
            dongs.append(address)
            dongs_count.append(count)
```

```python
plt.figure(figsize=(15, 10))  # 전체 그래프 캔버스 크기조절
plt.title('마포구 각 동별 cctv 개수', pad=30, fontsize=20)
plt.barh(dongs, dongs_count)
plt.show()
```

### 설치목적별 cctv 비율 Pie로 시각화하기

Figure_3.png

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
read_lines = []
with open('마포구_cctv.csv') as f:
    reader = csv.reader(f)
    for line in reader:
        read_lines.append(line)

all_purposes = []
for line in read_lines[1:]:
        if line[3]:  # 설치목적이 들어있는 인덱스 3
            purpose = line[3]
            all_purposes.append(purpose)
purposes = [] # x축에 넣을 고유한 설치목적 리스트
purposes_count = []  # 전체자료에서 해당 설치목적의 개수
etc = 0  # 너무 작은 카테고리는 기타로 함쳐서 표현하기 위해 사용된 변수
set_purposes = list(set(all_purposes))
for p in set_purposes:
    count = all_purposes.count(p)
    if count < 50:  # 개수가 50개 미만인 항목은 기타로 통합
        etc += count
    else:
        purposes.append(p)
        purposes_count.append(count)

purposes.append('기타')  # 기타항목을 새로 추가하고
purposes_count.append(etc)  # 누적된 etc값을 생성
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.figure(figsize=(7, 7))
plt.title('설치목적별 cctv 비율')
plt.pie(purposes_count, labels=purposes, autopct='%0.1f%%', wedgeprops=wedgeprops)
plt.show()
```

## 위도와 경도를 이용하여 설치된 위치 Scatter로 시각화하기

Figure_4.png

image.png

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
read_lines = []
with open('마포구_cctv.csv') as f:
    reader = csv.reader(f)
    for line in reader:
        read_lines.append(line)

x = []  # 경도값을 담을 x축 리스트
y = []  # 위도값을 담을 y축 리스트
for line in read_lines[1:]:
        if line[11]:
            x.append(float(line[11]))  # 경도가 들어있는 인덱스 11
            # csv파일의 위도와 경도는 문자열 -> float으로 숫자형 변환이 필요함
        if line[10]:
            y.append(float(line[10]))  # 위도가 들어있는 인덱스 10

plt.scatter(x, y, alpha=0.5)
plt.show()
```
