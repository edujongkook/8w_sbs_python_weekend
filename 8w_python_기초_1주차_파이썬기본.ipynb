## 주석 (comment)

프로그램 코드 문서화 (docstring)
- 코드들의 최상단에 """ 내용 """ 을 이용하여 문서작업을 하면 주석처리와 동시에 정규적인 문서내용으로 기록됨

✅ p29 예제

```python
import math  # 모듈 import <- 설명 정보를 입력하실수 있어요

def get_area(radius):
    """
    개요: 반지름을 전달하면 원의 넓이를 반환하는 get_area() 함수
    작성자 : 홍길동
    최초작성일 : 2023-10-07
    """
    area = math.pi * math.pow(radius, 2)  # 변수에 계산 값 대입
    return area  # 함수 결과값 반환
```

### 한줄 주석
- 해시태그 (#) 뒤에 오는 내용은 프로그램 실행 시 제외

```python
radius = 1.5
area = get_area(radius)  # get_area() 함수 호출 결과를 area 변수에 저장
print(area)
print(get_area.__doc__)  # Docstring 속성
```

```python
# 고정된 가로, 세로, 높이 값을 저장
box_size_limit = (100, 200, 500)
print(box_size_limit)
```

```python
box_size_limit[1] = 300  # 수정하려고 하면 에러발생
```

#### 튜플 순환해보기
마찬가지로 반복문을 이용해서 튜플 항목에 순서대로 접근할 수 있습니다.

```python
for dimension in box_size_limit:
    print(dimension)
```

### 딕셔너리 (dictionary)
키(key)와 값(value)이 서로 연결되어 함께 저장되는 데이터 형태 입니다


```python
user_mario = {'name' : 'mario',
              'age' : 25,
              'point': 30000}
print(user_mario)
```

인덱스가 아닌 키를 사용해서 값에 접근할 수 있습니다.

```python
user_mario['name']
```

#### f-string 사용 시 조심
(키 값이 문자열로 되어 있다면 따옴표를 사용하기 때문에)

```python
print(f"user name : {user_mario['name']}, point : {user_mario['point']}")
```

#### 값(value)의 수정, 추가와 삭제

```python
favorite_sports = {
    'kim' : 'football',
    'smith' : 'tennis',
    'park' : 'swimming'
}
```

```python
favorite_sports['kim'] = 'golf'  # 기존의 키를 사용해 새로운 값으로 수정가능
print(favorite_sports)
```

```python
favorite_sports['son'] = 'football'  # 새로운 키와 값으로 추가
print(favorite_sports)
```

```python
del favorite_sports['park']  # del 을 사용, 키값으로 키-값 전체 삭제
print(favorite_sports)
```

### 딕셔너리의 순환  
기본적으로는 키의 정보만 순환됩니다.

```python
for key in favorite_sports:
    print(key)
```

### ⚡ mutable / immutable -p47

- mutable 은 수정해도 저장된 위치(메모리 주소)가 바뀌지 않음 (list / set / dict)

```python
ex_list = [1, 2, 3 ,4 ,5]
id(ex_list)  # id() 는 변수의 메모리 주소를 표시
```

```python
ex_list.append(6)
id(ex_list)
```

- immutable 은 값에 손을 대는 순간
저장되는 위치가 새로운 곳으로 바뀜 (int / float / str / tuple)

```python
s = "hello"
id(s)
```

```python
s = "world"
id(s)
```

- mutable 변수를 대입하는 경우 같은 위치를 가르키는 두개의 변수가 됨
따라서 수정사항도 함께 공유

```python
ex_list_new = ex_list
```

```python
id(ex_list)
```

```python
id(ex_list_new)
```

#### ⚡ 새로운 저장공간에 복사하고 싶은 경우 copy 모듈 사용

```python
import copy
ex_list_new = copy.deepcopy(ex_list)
# 새로운 공간을 만들어 값을 복사하는 것을 deepcopy (깊은 복사)
# 저장되있는 장소의 주소만 복사하는것을 Shallow copy(얕은 복사) 라고 함
```

```python
id(ex_list)
```

```python
id(ex_list_new)
```
